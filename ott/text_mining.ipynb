```python
import pandas as pd
from ast import literal_eval

df = pd.read_csv('./test_TM.csv')
df['zsl'].tolist()
```

```python
container = []
for df_zsl in df['zsl'].tolist():
    dict_zsl= literal_eval(df_zsl)
    casting = dict_zsl['출연진']
    acting = dict_zsl['연기력']
    scenario = dict_zsl['시나리오']
    direction = dict_zsl['연출']
    character = dict_zsl['등장인물']
    context = [casting, acting, scenario, direction, character]
    container.append(context)
container
```

```python
df_zsl = pd.DataFrame(container, columns=['출연진', '연기력', '시나리오', '연출', '등장인물'])
```

```python
df_score = df['score']
```

```python
df_preprocessed = pd.concat([df_zsl, df_score], axis=1)
df_preprocessed
```

```python
df_preprocessed.to_csv('test_preprocessed.csv')
```
